```python
print('Hostname:', socket.gethostname())
if 'samuel' in socket.gethostname().lower():
    path_to_data='../data'
else:
    path_to_data='/scratch/spf248/covid/data'
    
path_to_shp=os.path.join(path_to_data,'shp')
country_gis='USA'
level_gis=2
url_root='https://biogeo.ucdavis.edu/data/gadm3.6/shp'
print('Country:', country_gis)
```

```python
os.path.join(url_root,'gadm36_'+country_gis+'_shp.zip')
```

```python
if os.path.exists(os.path.join(path_to_shp,'gadm36_'+country_gis+'_shp.zip')):
    print('Shapefiles exist')
else:
    start = timer()
    print('Download shapefiles')
    filename=wget.download(os.path.join(url_root,'gadm36_'+country_gis+'_shp.zip'),out=path_to_shp)
    with zipfile.ZipFile(filename, 'r') as zip_ref:
        zip_ref.extractall(filename.replace('.zip',''))
    print("Done in", round(timer()-start), "sec")
```
